```python
#Librosa is a Python package for audio and music analysis, right? Because we are doing speech to text, so we need this. So it provides various functions to quickly extract key audio features and metrics from the audio files
import librosa
import torch

import IPython.display as display

from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer

import numpy as np
```

```python
#facebook
tokenizer= Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-base-960h")
model= Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-base-960h")
```

```python
audio, sampling_rate = librosa.load("/content/audio.mp4a.wav" , sr=16000)
```

```python
audio, sampling_rate
```

```python
display.Audio("/content/audio.mp4a.wav", autoplay=True)
```

```python
input_values = tokenizer(audio, return_tensors= 'pt').input_values
input_values
```

```python
#logits means non-normalized predictions
logits = model(input_values).logits
logits
```

```python
predicted_ids= torch.argmax(logits, dim = -1)
```

```python
transcriptions = tokenizer.decode(predicted_ids[0])
```
